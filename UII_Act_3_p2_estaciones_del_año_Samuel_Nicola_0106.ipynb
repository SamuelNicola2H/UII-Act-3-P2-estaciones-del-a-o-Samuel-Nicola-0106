#### Redaccion del problema:Escribe un software para manejar una lista de estaciones del año
#### menu
1.  Agregar
2.  Modificar
3.  Borrar
4.  Ver listas
5.  Salir



```python
Estaciones = []

while True:
  print("1 Agregar elemento a la lista")
  print("2 Modificar elemento de la lista")
  print("3 Borrar elemento de la lista")
  print("4 Mostrar elementos de la lista ")
  print("5 Salir")

  Opcion = int(input("Inserte la opcion a realizar: "))

  if Opcion == 1:
    Nombre = input("Inserte el nombre de la estación que desea agregar: ")
    Estaciones.append(Nombre)

  elif Opcion == 2:
    Posicion = int(input("Inserte la posicion de la estación que desea modificar: "))
    Nombre = input("Inserte el nuevo nombre de la estación: ")
    Estaciones[Posicion] = Nombre

  elif Opcion == 3:
    Posicion = int(input("Inserte la posicion de la estación que desea borrar: "))
    Estaciones.pop(Posicion)

  elif Opcion == 4:
    for X in Estaciones:
      print(X)

  elif Opcion == 5:
    break

  else:
    print("Opcion no valida")

print("programa realizado por Samuel Nicola NL  = 40")
```
